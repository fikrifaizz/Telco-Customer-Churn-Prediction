```python
gender_churn = data.groupby('gender')['Churn'].apply(lambda s: (s=='Yes').sum()).reset_index(name='Total Churn')
gender_churn.columns = ['Gender', 'Total Churn']
print(gender_churn)
plt.figure(figsize=(10, 6))
gender_churn['Total Churn'].plot(kind='bar', color='blue', edgecolor='black')
plt.title('Total Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Churn')
plt.xticks(ticks=np.arange(len(gender_churn['Gender'].astype(str).tolist())), labels=gender_churn['Gender'].astype(str).tolist(), rotation=0)
plt.tight_layout()
plt.show()
```

```python
online_backup_churn = data.groupby('OnlineBackup')['Churn'].apply(lambda s: (s=='Yes').sum()).reset_index(name='Total Churn')
online_backup_churn.columns = ['Online Backup', 'Total Churn']
online_backup_churn['Total Churn'] = sorted(online_backup_churn['Total Churn'], reverse=True)
print(online_backup_churn)
plt.figure(figsize=(10, 6))
online_backup_churn['Total Churn'].plot(kind='bar', color='green', edgecolor='black')
plt.title('Total Churn by Online Backup')
plt.xlabel('Online Backup')
plt.ylabel('Total Churn')
plt.xticks(ticks=np.arange(len(online_backup_churn['Online Backup'].astype(str).tolist())), labels=online_backup_churn['Online Backup'].astype(str).tolist(), rotation=0)
```

```python
dependent_churn = data.groupby('Dependents')['Churn'].apply(lambda s: (s=="Yes").sum()).reset_index(name='Total Churn')
dependent_churn.columns = ['Dependents', 'Total Churn']
dependent_churn['Total Churn'] = sorted(dependent_churn['Total Churn'], reverse=True)
print(dependent_churn)
plt.figure(figsize=(10, 6))
dependent_churn['Total Churn'].plot(kind='bar', color='red', edgecolor='black')
plt.xticks(ticks=np.arange(len(dependent_churn['Dependents'].astype(str).tolist())), labels=dependent_churn['Dependents'].astype(str).tolist(), rotation=0)
plt.title('Total Churn by Dependents')
plt.xlabel('Dependents')
```
